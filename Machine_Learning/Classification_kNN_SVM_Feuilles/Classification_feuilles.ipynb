# Classification de feuilles à l'aide du kNN puis d'une SVM

On cherche à classifier des feuilles d'une centaine d'espèces en fonction de 200 caractéristiques physiques environ.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv('Dataset_feuilles_1.csv')
```

```python
raw_data.head()
```

```python
raw_data.shape
```

```python
# Recherche du nombre d'espèces dans le training set
dt_sorted = np.sort(raw_data['species'].unique())
print("Inventaire des espèces de feuilles : {}".format(dt_sorted))
```

```python
print("Nombre d'espèces de feuilles : {}".format(dt_sorted.size))
```

```python
# Sélection des données en entrée et des étiquettes
X = raw_data.iloc[:,2:194]
y = raw_data.iloc[:,1]
```

```python
y.head()
```

```python
# Séparation du dataset en training et test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
# Vérification des formes des données
X_train.shape
```

```python
X_test.shape
```

```python
# Affichage des espèces présentes dans le training set ainsi que leur nombre
dt_train = np.sort(np.unique(y_train))
print("Inventaire des espèces de feuilles : {}".format(dt_train))
print("Nombre d'espèces de feuilles : {}".format(dt_train.size))
```

```python
# Affichage des espèces présentes dans le test set ainsi que leur nombre
dt_test = np.sort(np.unique(y_test))
print("Inventaire des espèces de feuilles : {}".format(dt_test))
print("Nombre d'espèces de feuilles : {}".format(dt_test.size))
```

### Baseline avec un kNN

```python

from sklearn import neighbors

errors = []
for k in range(2,10):
    knn = neighbors.KNeighborsClassifier(k)
    errors.append(100*(1-knn.fit(X_train, y_train).score(X_test, y_test)))
plt.plot(range(2,10), errors, 'o-')
plt.show()
```

```python
# choix d'un nombre de voisins puis entraînrement du kNN
knn = neighbors.KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
print("accuracy : {}".format(knn.score(X_test, y_test)))
```

```python
from sklearn import metrics
y_pred = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
# Nombre d'espèces prédites dans l'échantillon de test :
np.unique(y_pred).size
```

```python
# Nombre d'espèces effectives dans l'échantillon de test :
np.unique(y_test).size
```

### Optimisation avec une SVM et utilisation d'une GridSearch

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
```

```python
# utilisation d'un SVM multiclasse OVR avec régularisation ridge et 10 valeurs de C testées
svm_ovr = LinearSVC(penalty='l2', dual=False, multi_class="ovr", max_iter=1000)
params = {'C': np.logspace(-3,3,10)}
gsvm_ovr = GridSearchCV(svm_ovr, params, cv=10)
gsvm_ovr.fit(X_train, y_train)
print(gsvm_ovr.best_params_)
y_pred = gsvm_ovr.predict(X_test)
print("Précision : {}".format(metrics.accuracy_score(y_test, y_pred)))
```

```python
# pour tester manuellement un résultat réel avec un résultat prédit :
i = 124
print("Espèce réelle : {}".format(y_test.iloc[i]))
print("Espèce prédite : {}".format(y_pred[i]))
```

```python
# pour montrer automatiquement les erreurs :
erreurs = list()
for i in range(y_test.size):
    if (y_test.iloc[i] != y_pred[i]):
        erreurs.append(i)

reelles = []
predites = []
for i in erreurs:
    reelles.append(y_pred[i])
    predites.append(y_test.iloc[i])

d = {'réelles': reelles,
     'prédites': predites}
d_df = pd.DataFrame(d)
d_df
```

```python
# utilisation d'un SVM multiclasse OVR avec régularisation ridge et 30 valeurs de C testées
svm_ovr = LinearSVC(penalty='l2', dual=False, multi_class="ovr", max_iter=1000)
params = {'C': np.logspace(-3,3, 30)}
gsvm_ovr = GridSearchCV(svm_ovr, params, cv=10)
gsvm_ovr.fit(X_train, y_train)
print(gsvm_ovr.best_params_)
y_pred = gsvm_ovr.predict(X_test)
print("Précision : {}".format(metrics.accuracy_score(y_test, y_pred)))
```

```python
# utilisation d'un SVM multiclasse OVR avec régularisation lasso et 10 valeurs de C testées
svm_ovr = LinearSVC(penalty='l1', dual=False, multi_class="ovr", max_iter=1000)
params = {'C': np.logspace(-3,3, 10)}
gsvm_ovr = GridSearchCV(svm_ovr, params, cv=10)
gsvm_ovr.fit(X_train, y_train)
print(gsvm_ovr.best_params_)
y_pred = gsvm_ovr.predict(X_test)
print("Précision : {}".format(metrics.accuracy_score(y_test, y_pred)))
```

```python
# utilisation du SVM multiclasse "crammer_singer" avec régularisation ridge et 10 valeurs de C testées
svm_crammer = LinearSVC(penalty='l2', dual=False, multi_class="crammer_singer", max_iter=1000)
params = {'C': np.logspace(-3,3, 10)}
gsvm_crammer = GridSearchCV(svm_crammer, params, cv=10)
gsvm_crammer.fit(X_train, y_train)
print(gsvm_crammer.best_params_)
y_pred = gsvm_crammer.predict(X_test)
print("Précision : {}".format(metrics.accuracy_score(y_test, y_pred)))
```
